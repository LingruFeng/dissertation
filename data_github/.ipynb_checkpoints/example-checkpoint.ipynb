## 以需求点到供给点的距离作为cost

```python
B = access(demand_df =chi_pop,demand_index='geoid',demand_value='pop',supply_df= chi_doc,supply_index= 'geoid',supply_value=['doc','dentist']
           )
B.demand_df = B.demand_df.to_crs(epsg = 3528)
B.supply_df = B.supply_df.to_crs(epsg = 3528)
B.create_euclidean_distance(name= 'euclidean_neighbors', threshold = 250000)
B.cost_df
```

```python
c=B.cost_df
```

# 创建一个以距离为cost的dataframe用于可达性计算

```python
A = access(demand_df            = chi_pop,
           demand_index         = 'geoid',
           demand_value         = 'pop',
           supply_df            = chi_doc,
           supply_index         = 'geoid',
           supply_value         = ['doc','dentist'],
           cost_df              = c,
           cost_origin          = 'origin',
           cost_dest            = 'dest',
           cost_name            = 'euclidean_neighbors',
           neighbor_cost_df     = c,
           neighbor_cost_origin = 'origin',
           neighbor_cost_dest   = 'dest',
           neighbor_cost_name   = 'euclidean_neighbors')
```

```python
A.cost_df
```

## 结果

```python
A.two_stage_fca(name = "2sfca5",max_cost = 1e5)
A.two_stage_fca(name = "2sfca4",max_cost = 1e4)
```

# 根据权重将两个指标合为一个标准化的可达性数值

```python
A.score(name = "2sfca4_combo", col_dict = {"2sfca4_doc" : 0.8, "2sfca4_dentist" : 0.2})
```

```python
A.score(name = "2sfca5_combo", col_dict = {"2sfca5_doc" : 0.8, "2sfca5_dentist" : 0.2})
```

```python
il_map = chi_doc.set_index('geoid')[["geometry"]].join(A.access_df, how = "inner")

def plot(v, ax = None, leg = True):
    
    swap_colors = "raam" in v
    
    il_map.plot(v, legend = leg, figsize = (7, 7), 
                cmap = "viridis_r" if swap_colors else "viridis", 
                markersize = 5, alpha = 0.8, ax = ax,
                vmin = il_map[v].quantile(0.05), vmax = il_map[v].quantile(0.95)).set_axis_off()
```
